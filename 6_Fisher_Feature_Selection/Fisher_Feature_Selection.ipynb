## Relation between Fisher Criterion and Least Squares

### Proof of $w_0=-\frac{\mathbf{w}^T\sum_{i=1}^n(\mathbf{x}_i)}{n}=-\mathbf{w}^T\mathbf{m}$

According to the definition of the error of the square loss:
$$\begin{cases}
\frac{\partial E}{\partial \mathbf{w}}=\sum_{i=1}^n((\mathbf{w}^T\mathbf{x}_i+w_0-t_i)\mathbf{x}_i^T)\\
\frac{\partial E}{\partial w_0}=\sum_{i=1}^n(\mathbf{w}^T\mathbf{x}_i+w_0-t_i)
\end{cases}$$

Therefore:

$$\begin{cases}
\frac{\partial E}{\partial \mathbf{w}}=
\mathbf{w}^T\sum_{i=1}^n(\mathbf{x}_i\mathbf{x}_i^T)+w_0\sum_{i=1}^n(\mathbf{x}_i^T)-\sum_{i=1}^n(t_i\mathbf{x}_i^T)\ =0\\
\frac{\partial E}{\partial w_0}=\mathbf{w}^T\sum_{i=1}^n(\mathbf{x}_i)+nw_0-\sum_{i=1}^n(t_i)=0
\end{cases}$$

For the reason that $t_1=\frac{n}{n_1}, t_2=-\frac{n}{n_2}$, we can get $\sum_{i=1}^n(t_i)=0$

Therefore, we get:

$$w_0=-\frac{\mathbf{w}^T\sum_{i=1}^n(\mathbf{x}_i)}{n}=-\mathbf{w}^T\mathbf{m}$$

### Proof of $(S_w+\frac{n_1n_2}{n}S_B)\mathbf{w}=n(\mathbf{m_1-m_2})$

$$
\mathbf{w}^T\sum_{i=1}^n(\mathbf{x}_i\mathbf{x}_i^T)-\mathbf{w}^T\mathbf{m}\sum_{i=1}^n(\mathbf{x}_i^T)-\sum_{i=1}^n(t_i\mathbf{x}_i^T)=0\Rightarrow
\mathbf{w}^T\sum_{i=1}^n((\mathbf{x}_i-\mathbf{m})\mathbf{x}_i^T)=\sum_{i=1}^n(t_i\mathbf{x}_i^T)$$

$$\sum_{i=1}^n(t_i\mathbf{x}_i^T)=\frac{n}{n_1}\sum_{\mathcal{C}_1}(\mathbf{x}_i^T)-\frac{n}{n_2}\sum_{\mathcal{C}_2}(\mathbf{x}_i^T)=n(\mathbf{m_1-m_2}), \ \ \sum_{i=1}^n(\mathbf{x}_i(\mathbf{x}_i-\mathbf{m})^T)\mathbf{w}=n(\mathbf{m_1-m_2})$$

Now, we are about to solve the parameter $\sum_{i=1}^n(\mathbf{x}_i(\mathbf{x}_i-\mathbf{m})^T)$ as following:

$$\begin{aligned}
&\sum_{i=1}^n(\mathbf{x}_i(\mathbf{x}_i-\mathbf{m})^T)=
\sum_{\mathcal{C}_1}(\mathbf{x}_i(\mathbf{x}_i-\frac{n_1\mathbf{m_1}+n_2\mathbf{m_2}}{n})^T)
+\sum_{\mathcal{C}_2}(\mathbf{x}_i(\mathbf{x}_i-\frac{n_1\mathbf{m_1}+n_2\mathbf{m_2}}{n})^T)\\
&=	\frac{n_1}{n}\sum_{\mathcal{C}_1}(\mathbf{x}_i(\mathbf{x}_i-\mathbf{m_1})^T)
  	+\frac{n_2}{n}\sum_{\mathcal{C}_1}(\mathbf{x}_i(\mathbf{x}_i-\mathbf{m_2})^T)
	+\frac{n_1}{n}\sum_{\mathcal{C}_2}(\mathbf{x}_i(\mathbf{x}_i-\mathbf{m_1})^T)
  	+\frac{n_2}{n}\sum_{\mathcal{C}_2}(\mathbf{x}_i(\mathbf{x}_i-\mathbf{m_2})^T)\\
&=	\frac{n_1}{n}\sum_{\mathcal{C}_1}((\mathbf{x}_i-\mathbf{m_1})(\mathbf{x}_i-\mathbf{m_1})^T)
	+\frac{n_1}{n}\sum_{\mathcal{C}_1}(\mathbf{m_1}(\mathbf{x}_i-\mathbf{m_1})^T)
	+\frac{n_2}{n}\sum_{\mathcal{C}_1}((\mathbf{x}_i-\mathbf{m_1})(\mathbf{x}_i-\mathbf{m_2})^T)
	+\frac{n_2}{n}\sum_{\mathcal{C}_1}(\mathbf{m_1}(\mathbf{x}_i-\mathbf{m_2})^T)\\
&	+\frac{n_1}{n}\sum_{\mathcal{C}_2}((\mathbf{x}_i-\mathbf{m_2})(\mathbf{x}_i-\mathbf{m_1})^T)
	+\frac{n_1}{n}\sum_{\mathcal{C}_2}(\mathbf{m_2}(\mathbf{x}_i-\mathbf{m_1})^T)
	+\frac{n_2}{n}\sum_{\mathcal{C}_2}((\mathbf{x}_i-\mathbf{m_2})(\mathbf{x}_i-\mathbf{m_2})^T)
	+\frac{n_2}{n}\sum_{\mathcal{C}_2}(\mathbf{m_2}(\mathbf{x}_i-\mathbf{m_2})^T)\\
&=	\frac{n_1}{n}\sum_{\mathcal{C}_1}((\mathbf{x}_i-\mathbf{m_1})(\mathbf{x}_i-\mathbf{m_1})^T)
	+\frac{n_2}{n}\sum_{\mathcal{C}_2}((\mathbf{x}_i-\mathbf{m_2})(\mathbf{x}_i-\mathbf{m_2})^T)
	+\frac{n_2n_1}{n}\mathbf{m_1}(\mathbf{m_1}-\mathbf{m_2})^T
	-\frac{n_1n_2}{n}\mathbf{m_2}(\mathbf{m_1}-\mathbf{m_2})^T\\
&	+\frac{n_2}{n}\sum_{\mathcal{C}_1}((\mathbf{x}_i-\mathbf{m_1})(\mathbf{x}_i-\mathbf{m_1})^T)
	+\frac{n_2}{n}\sum_{\mathcal{C}_1}((\mathbf{x}_i-\mathbf{m_1})(\mathbf{m_1}-\mathbf{m_2})^T)\\
&	+\frac{n_1}{n}\sum_{\mathcal{C}_2}((\mathbf{x}_i-\mathbf{m_2})(\mathbf{x}_i-\mathbf{m_2})^T)
	+\frac{n_1}{n}\sum_{\mathcal{C}_2}((\mathbf{x}_i-\mathbf{m_2})(\mathbf{m_2})^T-\mathbf{m_1})^T)\\
&=	S_w+\frac{n_1n_2}{n}S_B
\end{aligned}$$

Finally we got:

$$(S_w+\frac{n_1n_2}{n}S_B)\mathbf{w}=n(\mathbf{m_1-m_2})$$

### Proof of $w \propto S_w^{-1}(\mathbf m_2 - \mathbf m_1)$

In order to prove $\mathbf{w}=kS_w^{-1}(\mathbf{m_2-m_1})$, we have to prove:

$$k\frac{n_1n_2}{n}S_BS_w^{-1}(\mathbf{m_2-m_1})=(n+k)(\mathbf{m_1-m_2})$$

That is to say:

$$S_BS_w^{-1}(\mathbf{m_2-m_1})=(n+k+k\frac{n_1n_2}{n})(\mathbf{m_2-m_1})$$

Which means that $S_BS_w^{-1}$ have an eigenvector $\mathbf{m_2-m_1}$, and the eigenvalue of this vector is not $n$

According to the basic knowledgement:

$$S_BS_w^{-1}(\mathbf{m_2-m_1})=(\mathbf{m_2-m_1})(\mathbf{m_2-m_1})^TS_w^{-1}(\mathbf{m_2-m_1})\\
=(\mathbf{m_2-m_1})((\mathbf{m_2-m_1})^TS_w^{-1}(\mathbf{m_2-m_1}))$$

It is obvious that the $((\mathbf{m_2-m_1})^TS_w^{-1}(\mathbf{m_2-m_1}))$ is a scalar and in the most case, $((\mathbf{m_2-m_1})^TS_w^{-1}(\mathbf{m_2-m_1})) \ne n$

therefore, we conclude that the $\mathbf{m_2-m_1}$ is the eigen vector of $S_BS_w^{-1}$ and 

$$w \propto S_w^{-1}(\mathbf m_2 - \mathbf m_1)$$

## Fisher's discriminant for multiple classes

### Decompose the covariance matrix

$$\begin{aligned}
S_T &= \sum_{n=1}^N(\mathbf x_n - \sum_{k=1}^K\frac{N_k}{N}\mathbf m_k)(\mathbf x_n - \sum_{k=1}^K\frac{N_k}{N}\mathbf m_k)^{\mathrm T} = \sum_{j=1}^K\sum_{n\in\mathcal C_j}(\mathbf x_n - \sum_{k=1}^K\frac{N_k}{N}\mathbf m_k)(\mathbf x_n - \sum_{k=1}^K\frac{N_k}{N}\mathbf m_k)^{\mathrm T}\\
&= \sum_{j=1}^K\sum_{n\in\mathcal C_j}(\mathbf x_n - \mathbf m_j + \mathbf m_j  - \sum_{k=1}^K\frac{N_k}{N}\mathbf m_k)(\mathbf x_n - \mathbf m_j + \mathbf m_j  - \sum_{k=1}^K\frac{N_k}{N}\mathbf m_k)^{\mathrm T}\\
&= \sum_{j=1}^K\sum_{n\in\mathcal C_j}(\mathbf x_n - \mathbf m_j)(\mathbf x_n - \mathbf m_j)^{\mathrm T} + (\mathbf m_j  - \sum_{k=1}^K\frac{N_k}{N}\mathbf m_k)(\mathbf m_j  - \sum_{k=1}^K\frac{N_k}{N}\mathbf m_k)^{\mathrm T}  \\
&+ (\mathbf x_n - \mathbf m_j)(\mathbf m_j  - \sum_{k=1}^K\frac{N_k}{N}\mathbf m_k)^{\mathrm T} + (\mathbf x_n - \mathbf m_j)^{\mathrm T}(\mathbf m_j  - \sum_{k=1}^K\frac{N_k}{N}\mathbf m_k) \text{    //It is obvious that these two items are 0}\\
&= S_W + \sum_{j=1}^KN_j(\mathbf m_j  - \sum_{k=1}^K\frac{N_k}{N}\mathbf m_k)(\mathbf m_j  - \sum_{k=1}^K\frac{N_k}{N}\mathbf m_k)^{\mathrm T} = S_W + \sum_{k=1}^KN_k(\mathbf m_k - \mathbf m)(\mathbf m_k - \mathbf m)^{\mathrm T} = S_W + S_B
\end{aligned}$$

### $s_W$ and $s_B$ in the projected space

It is obvious that $s_B = \mathbf W^{\mathrm T}S_B\mathbf W$, and $s_W = \mathbf W^{\mathrm T}S_W\mathbf W$

### $J(\mathbf W)$
In short, we can got $$J(\mathbf W) = \frac{\prod_{diag} \mathbf W^{\mathrm T}S_B\mathbf W}{\prod_{diag}\mathbf W^{\mathrm T}S_W\mathbf W}$$

Setting $\mathbf W = \begin{pmatrix} w_1 & w_2 & \cdots & w_{D'}\end{pmatrix}$, we got:

$$\prod_{diag} \mathbf W^{\mathrm T}S_B\mathbf W = \prod_i w_i^{\mathrm T}S_Bw_i, \ \ \prod_{diag} \mathbf W^{\mathrm T}S_W\mathbf W = \prod_i w_i^{\mathrm T}S_Ww_i$$

Therefore:

$$J(\mathbf W) = \frac{\prod_i w_i^{\mathrm T}S_Bw_i}{\prod_i w_i^{\mathrm T}S_Ww_i}$$

### The optimizating method

It is obvious that the $S_W$ and $S_B$ could be diagonalized since they are symmetric matrices. Therefore, it is so obvious that by choosing the eigen vector of matrix $S_W^{-1}S_B$, the $J(\mathbf W)$ would get its maximum.

That is to say, setting the eigen value of $S_W^{-1}S_B$ is $\lambda_1 > \lambda_2 > \cdots,\lambda_D \ge 0$, we got 

$$\lambda_i w_i = S_W^{-1}S_Bw_i$$

> NOTICE that all of the eigen value is greater than 0 because the $S_W,S_B$ are postive definition matrix

### Maxium of $D'$

It is obvious that the rank of $S_W$ is $D$ in most cases, however, $S_B = \sum N_k\Delta_m\Delta_m^{\mathrm T}$. $Rank(\Delta_m\Delta_m^{\mathrm T}) \le Rank(\Delta_m) = 1$, therefore, $Rank (S_B) \le \sum Rank(\Delta_m\Delta_m^{\mathrm T}) \le k$. Therefore, $S_B,SW^{-1}S_B$ could have $k$ non-zero eigen value at most.

If we choose a vector which have a zero eigen value for $SW^{-1}S_B$, it is obvious that the $J(\mathbf W)$ will drop to 0$.

Therefore, we are able to find at most $K$ 'features'

## Programming

### Summarize the equations

$$J = \frac1n\sum_{i=1}^n(\frac12(y_i-\mathbf w^\mathrm T\psi(\mathbf x_i))^2 + \lambda \|\mathbf w\|_1$$

$$a_k = \frac1n\sum_{i=1}^n\psi_k^2(\mathbf x_i)\ \ \ \ c_k = \frac1n\sum_{i=1}^n\psi_k(\mathbf x_i)(y_i-\mathbf w_{-k}^{\mathrm T}\psi_{-k}(\mathbf x_i))$$

$$\frac{\partial J}{\partial w_k} = \begin{cases}
\{a_kw_k - c_k - \lambda\},\ \ w_k < 0\\
[-c_k-\lambda,-c_k+\lambda], \ \ w_k = 0\\
\{a_kw_k - c_k + \lambda\},\ \ w_k > 0
\end{cases}$$

### interpret the coefficient $c_k$

The $c_k$ is calculated by $\frac\partial{\partial w_k}\frac1n\sum_{i=1}^n(\frac12(y_i-\mathbf w^\mathrm T\psi(\mathbf x_i))^2$ as following:

$$\begin{aligned}
&\frac\partial{\partial w_k}\frac1n\sum_{i=1}^n(\frac12(y_i-\mathbf w^\mathrm T\psi(\mathbf x_i))^2 = \frac\partial{\partial w_k}\frac1n\sum_{i=1}^n(\frac12(y_i-\mathbf w_{-k}^{\mathrm T}\psi_{-k}(\mathbf x_i) - w_k \psi_k(\mathbf x_i))^2 \\ 
=&\frac1n\sum_{i=1}^n(-\psi_k(\mathbf x_i)(y_i-\mathbf w_{-k}^{\mathrm T}\psi_{-k}(\mathbf x_i) - w_k \psi_k(\mathbf x_i)) = \frac1n\sum_{i=1}^n\psi_k^2(\mathbf x_i)w_k - \frac1n\sum_{i=1}\psi_k(\mathbf x_i)(y_i-\mathbf w_{-k}^{\mathrm T}\psi_{-k}(\mathbf x_i)) \\
=&a_kw_k - c_k
\end{aligned}$$

Therefore, $c_k = \frac1n\sum_{i=1}^n\psi_k(\mathbf x_i)(y_i-\mathbf w_{-k}^{\mathrm T}\psi_{-k}(\mathbf x_i))$

The $c_k$ should be interpreted as the "significance" of each features, that is to say, if the function can fit the point well without $w_k$, i.e. $\forall i, \ y_i-\mathbf w_{-k}^{\mathrm T}\psi_{-k}(\mathbf x_i)\approx 0$, then the $c_k$ would be a relatively small value. Also, a feature with greater $\psi(\mathbf x_i)$ tend to have a greater significant, which means the greater $\psi(\mathbf x_i)$ gives more chance for the feature $k$ to fix the error of the original model $y_i-\mathbf w_{-k}^{\mathrm T}\psi_{-k}(\mathbf x_i)$ with a relatively small $w_k$

### Solve the non-smooth optimality condition

$$\frac{\partial J}{\partial w_k} = \begin{cases}
\{a_kw_k - c_k - \lambda\},\ \ w_k < 0\\
[-c_k-\lambda,-c_k+\lambda], \ \ w_k = 0\\
\{a_kw_k - c_k + \lambda\},\ \ w_k > 0
\end{cases}$$

When $c_k < -\lambda$, ONLY function $a_kw_k - c_k - \lambda, w_k < 0$ will have a root as zero $\hat w_k = \frac{c_k + \lambda}{a_k} < 0$

For the same reason, the zero-crossing $\hat w_k$ should be described as the following function:

$$\hat w_k = \begin{cases}
\frac{c_k + \lambda}{a_k},\ \ c_k < -\lambda\\
0, \ \ c_k \in [-\lambda,+\lambda]\\
\frac{c_k - \lambda}{a_k},\ \ c_k > +\lambda
\end{cases}$$

We can find out that the $\lambda$ is the limit of the "significance" of each feature: with the significance $|c_k| \le\lambda$, the feature would be ignored. Moreover, with a very important feature $|c_k| >> \lambda$, the $\hat w_k$ would be similar to the one without the L1-norm regularization.

The plot are shown as below:

```python
from matplotlib import pyplot as plt
plt.rcParams["figure.figsize"] = [12,4]
def partialJ(a,c,l):
    plt_bound = 3
    plt.plot([0,-plt_bound],[-c - l, -plt_bound * a - c - l])
    plt.plot([0,0],[-c - l,-c + l])
    plt.plot([0,plt_bound],[-c+l,plt_bound * a - c + l])
    if c < -l:
        plt.scatter((c+l)/a,0,label='zero crossing point')
    elif c > l:
        plt.scatter((c-l)/a,0,label='zero crossing point')
    else:
        plt.scatter(0,0,label='zero crossing point')
        
plt.subplot(1,3,1)
partialJ(1,-2,1)
plt.title('c_k < - lambda')
plt.legend()

plt.subplot(1,3,2)
partialJ(1,-0.5,1)
plt.title('- lambda < c_k < + lambda')
plt.legend()
plt.subplot(1,3,3)
partialJ(1,2,1)
plt.title('+ lambda < c_k')
plt.legend()
plt.show()
```
