# Canvas: Generate the ''Name Card'' Automatically

## Check out the Chinese Fonts and English Fonts

We use the System Fonts of Windows (both Chinese system and English System). In order to avoid bug characters, we should filter all of these fonts first, manually

## Generate the text automatically

```python
import os
from PIL import Image,ImageDraw,ImageFont
import requests
import numpy as np

font_name = './Fonts/origin/Deng.ttf'
font = ImageFont.truetype(font_name,24,)
url = 'http://a.hiphotos.baidu.com/image/pic/item/10dfa9ec8a136327f216788d9d8fa0ec09fac791.jpg'
response = requests.get(url)
with open('temp.png', 'wb') as file:  # 以byte形式将图片数据写入  
    file.write(response.content)  
    file.flush() 
img = Image.open('temp.png')
img = img.resize((300,200))
arr = np.mean(np.asarray(img),axis=(0,1))
draw = ImageDraw.Draw(img)
draw.text((0,50), u'你好,世界!', font=font, fill=tuple(255-arr.astype(int)))
img
```
